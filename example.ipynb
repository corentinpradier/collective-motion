# Utilisation

```python
from collective_motion import CollectiveMotion
import matplotlib.pyplot as plt
```

```python
cm = CollectiveMotion(
    time_tot=200,
    n_preys=300,
    n_predators=0,
    diffusion=0.8,
    couplage=0.9,
    radius_predators=10,
    radius_avoid=2,
    radius_influence=5,
    window_width=30,
    time_step=0.8,
)
```

```python
position, angle, order_parameter = cm.main()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(18, 7))

cm.plot_frame(position, angle, ax=ax[0], cmap="viridis", frame=0)
cm.plot_frame(position, angle, ax=ax[1], cmap="viridis", frame=25)
cm.plot_frame(position, angle, ax=ax[2], cmap="viridis", frame=100)
```

```python
cm.plot_run(position, angle, cmap="viridis")
```
